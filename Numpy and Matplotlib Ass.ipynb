```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y,'bo-',label='2x',linewidth=2)
plt.title('Line Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X-Axis')               
plt.ylabel('Y-Axis')               
plt.legend()
plt.show()
```

```python
x=['Java','Python','PHP','Javascript','C#','C++']
y=[17.6,22.2,8.8,8,7.7,6.7]
plt.bar(x,y)
plt.title('Popularity of various languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
x=['Java','Python','PHP','Javascript','C#','C++']
y=[17.6,22.2,8.8,8,7.7,6.7]
plt.pie(y,labels=x,autopct='%1.2f%%')
plt.title("Pie chart showing Popularity of Different Languages")
plt.show()
```
